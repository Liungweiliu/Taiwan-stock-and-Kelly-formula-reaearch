```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
```

```python
a = 0.5          #負獲利率(輸一次扣0.5元)  
b = 20           #正獲利率(贏一次拿回2+1元)
p = 0.3          #勝率
q = 1-p          #賠率(1-p)
```

```python
f = np.arange(0,1,0.01)              #資金分配比例
y = q*np.log(1-a*f)+p*np.log(1+b*f)  #期望報酬率
```

```python
myfont = fm.FontProperties(fname='C:/Windows/Fonts/msyh.ttc')
```

```python
#輸出中文
plt.rcParams['font.sans-serif']=['SimHei']
#最佳資金分配報酬
fig, ax = plt.subplots()

# ax = plt.gca()
ax.set_axis_bgcolor('#F3DFC1')
#ax.set
#orange curve 
plt.plot(f, y, '#FF8552', lw=3)
#red line
plt.plot(f, max(y)+f*0, 'r')
#draw the maximiun
plt.plot(f[np.argmax(y)-1], max(y), '*', color='#02FF24')
plt.xlabel("資金分配比例", fontproperties=myfont)
plt.ylabel('G(f)')
plt.title('Expected Value of Logarithm of Wealth vs. Bet as a Fraction of Wealth')
fig.text(0.85, 0.01, 'Kelly.F',
         fontsize=80, color='gray',
         ha='right', va='bottom', alpha=0.5)
plt.show()


print("最佳化投資比例可得%s投報率"%max(y))
```
